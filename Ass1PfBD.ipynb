```python
#Question 2
import math

def isPrimeNumber( n ):
        
    if n < 2: 
        return False
    if n == 2: 
        return True
    if n % 2 == 0: 
        return False
    
    r = math.sqrt( n )
    x = 3 
    while x <= r:
        if n % x == 0: 
            return False 
        x += 2 
        
    return True 

PrimeLst=[]
Sum=0
Ans=0
for i in range(1,1000000):
    if isPrimeNumber(i):
        PrimeLst.append(i)
for p in PrimeLst:
    Sum=Sum+p
    if Sum in PrimeLst:
        Ans=Sum  
print('Prime, below one-million, which can be written as the sum of the most consecutive primes is ',Ans)
```
